## Popularity Based Recommendation

- Popularity based recommendation system, shows the users, top n best products, for example top 250 movies on IMDB. These rankings are based on certain formula

```python
books_ratings=books.merge(ratings,on='ISBN')
books_ratings.head()
```

```python
reviews_summary=books_ratings.groupby('Book-Title')['Book-Rating'].agg(['count','mean']).reset_index()
reviews_summary
```

```python
# Now we will take those Books which has been reviewed at least 150 times.
t=reviews_summary[reviews_summary['count']>=150].index
```

```python
top_books=reviews_summary[reviews_summary['count']>=150]
```

```python
top_books=books_ratings.merge(top_books,on='Book-Title')
top_books.drop_duplicates(subset='Book-Title',inplace=True)
```

```python
top_books=top_books.sort_values(by='mean',ascending=False)
top_books=top_books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-S','Image-URL-M','Image-URL-L','count','mean']]
top_books.head()
```

## Collaborative Recommendation System

- So now we will implement collaborative recommendation system, for that first we will find out credible data.
- We will only take those users as input who has given atleast 200 reviews, and those books which are rated a certain number of times. This is done to make the recommendation credible, and users with only 3-4 reviews are not considered for recommendations or books rated lesser number of times are not recommended.

```python
users
```

```python
combined_data=books.merge(ratings,on='ISBN').merge(users,on='User-ID')
combined_data.head(2)
```

```python
combined_data.shape
```

```python
# Getting those rows where the users have rated atleast 200 times.

t=combined_data.groupby('User-ID').count()['Book-Title']>200
cleaned_user_data=combined_data[combined_data['User-ID'].isin(t[t].index.values)]
cleaned_user_data.shape
```

```python
# Now we are left with 500k rows. going forward we will filter out those books which were rated atleast 50 times.

x=cleaned_user_data.groupby('Book-Title').count()['ISBN']>=50
cleaned_user_books_data=cleaned_user_data[cleaned_user_data['Book-Title'].isin(x[x].index.values)]
cleaned_user_books_data
```

```python
cleaned_matrix=cleaned_user_books_data.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
cleaned_matrix
```

```python
cleaned_matrix=cleaned_matrix.fillna(0)
cleaned_matrix
```
